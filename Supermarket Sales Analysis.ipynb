```python
colname=df.select_dtypes(["int64","float64"]).columns
colname
```

```python
df[colname]
```

```python
from scipy.stats import skew
```

```python
for col in df[colname]:
    print(col)
    print(skew(df[col]))

    sns.distplot(df[col])
    plt.show()
```

```python
df.corr().style.background_gradient()
```

skew ----> Close to zero

Correlation---> Close to one

Conclusion:
    
1) Total is Highly Skewed and correlation not close to 1

2)Cogs is highly skewed and cogs not close to 1

3)Gross income is highly skewed and not correlation not close to 1

4)Tax 5% is highly skewed and not correlation not close to 1

Then use a log or sqrt are use to converted Skewed value close to zero

### sqrt

```python
df["Total"]=np.sqrt(df["Total"])
df["cogs"]=np.sqrt(df["cogs"])
df["gross income"]=np.sqrt(df["gross income"])
df["Tax 5%"]=np.sqrt(df["Tax 5%"])
```

```python
for col in df[colname]:
    print(col)
    print(skew(df[col]))

    sns.distplot(df[col])
    plt.show()
```
